```python
import cv2

image = cv2.imread('car.jpg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
invert_image_gray = 255 - image_gray
image_blur = cv2.GaussianBlur(invert_image_gray, (19,19), 0)
invert_image_blur = 255 - image_blur
image_sketch = cv2.divide(image_gray, invert_image_blur, scale = 255)
cv2.imwrite('sketch.png', image_sketch)
```
